# 🎬 Netflix Data Analysis
Analyze Netflix dataset to derive insights on genres, countries, and release trends.

## Objectives
- Analyze distribution of Movies and TV Shows
- Identify popular genres
- Explore release year and country trends

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
# Basic info
df.info()
df.isnull().sum()
```

```python
# Clean data
df.dropna(subset=['country', 'rating'], inplace=True)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
# Movies vs TV Shows
sns.countplot(x='type', data=df, palette='coolwarm')
plt.title('Movies vs TV Shows')
plt.show()
```

```python
# Top 10 Genres
df['listed_in'].value_counts().head(10).plot(kind='bar', color='lightblue')
plt.title('Top 10 Genres')
plt.show()
```

```python
# Release trend
df['release_year'].value_counts().sort_index().plot(kind='line', color='red')
plt.title('Content Release Trend')
plt.show()
```

```python
# Top 10 Countries
df['country'].value_counts().head(10).plot(kind='bar', color='green')
plt.title('Top Countries by Netflix Content')
plt.show()
```

## Insights
- Most content on Netflix is Movies
- Drama and Comedy are top genres
- USA and India lead in content production
